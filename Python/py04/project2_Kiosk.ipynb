```python
# 키오스크 클래스 (Project2)

class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]

    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')

    def menu_select(self):
        print()

        self.order_menu = []
        self.order_price = []
        self.price_sum = 0

        while True:
            n = int(input("음료 번호를 입력하세요 (주문 완료: 0) : "))

            if n == 0:
                break

            if n < 1 or n > len(self.menu):
                print("없는 메뉴입니다. 다시 주문해 주세요.")
                continue

            while True:
                t = int(input("HOT 음료는 1, ICE 음료는 2 : "))
                if t == 1:
                    temp = "HOT"
                    break
                elif t == 2:
                    temp = "ICE"
                    break
                else:
                    print("1과 2 중 하나를 입력하세요.\n")

            self.order_price.append(self.price[n - 1])
            self.order_menu.append(temp + ' ' + self.menu[n - 1])
            self.price_sum += self.price[n - 1]

            print(f"추가됨: {temp} {self.menu[n-1]} / {self.price[n-1]}원\n")

        print("주문 내역:", self.order_menu)
        print("가격 내역:", self.order_price)
        print("총 금액:", self.price_sum, "원")

    def pay(self):
        while True:
            means = int(input("지불 수단을 입력해주세요: (현금:1, 카드:2)\n"))
            self.means = means

            if self.means == 1:
                print('직원을 호출하겠습니다.')
                break
            elif self.means == 2:
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
                break
            else:
                print('다시 결제를 시도해 주세요.')

    def table(self):
        print()
        print('⟝' + '-' * 30 + '⟞')

        for _ in range(2):
            print('|' + ' ' * 31 + '|')

        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i], '원')

        print()
        print('합계 금액 :', sum(self.order_price), '원')

        for _ in range(2):
            print('|' + ' ' * 31 + '|')

        print('⟝' + '-' * 30 + '⟞')
```

```python
a = Kiosk()  # 객체 생성 
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
